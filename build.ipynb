```python
import glob
import os
import pandas as pd
from pathlib import Path
from natsort import natsorted
```

```python
list_txt=natsorted(list(glob.glob(os.path.join(".","policy-statement","*","data.txt"))))
```

```python
list_txt
```

```python
data={
    "n_cabinet":[],
    "text":[]
}
```

```python
for i in list_txt:
    n_cabinet=os.path.basename(os.path.dirname(i))
    with open(i,"r",encoding="utf-8-sig") as f:
        text='\n'.join([k.strip() for k in f.readlines()]).strip()
    data["n_cabinet"].append(int(n_cabinet))
    data["text"].append(text)
```

```python
df=pd.DataFrame.from_dict(data)
```

```python
df
```

```python
df.to_csv("data.csv",index=False)
```
